```python
#이미지 읽기
import cv2

imageFile = './bono.png' # '파일명,파일주소'
img  = cv2.imread(imageFile)    # 컬러
img2 = cv2.imread(imageFile, 0) # cv2.IMREAD_GRAYSCALE(흑백)
cv2.imshow('Lena color',img) #제목,파일명
cv2.imshow('Lena grayscale',img2) #제목,파일명
cv2.waitKey(0) # 0이면 key 입력이 있을때까지 무한대기
cv2.destroyAllWindows() #화면에 나타난 윈도우를 종료

```

```python
#이미지 저장(영상을 이미지로 저장할때에도 똑같이 적용)

import cv2

imageFile = './bono.jpg' #파일명
img = cv2.imread(imageFile) # cv2.imread(imageFile, cv2.IMREAD_COLOR)
cv2.imwrite('./bono.bmp', img) #bmp확장자로 저장
cv2.imwrite('./bono.png', img) #png확장자로 저장
cv2.imwrite('./bono.png',img, [cv2.IMWRITE_PNG_COMPRESSION, 9]) #압축률 0~9까지 기본값:3
cv2.imwrite('./bono.jpg', img, [cv2.IMWRITE_JPEG_QUALITY, 90]) #0~100까지 품질 설정가능 
```

```python
#BGR을 RGB로 바꾸기(cv2는 BGR형태로 저장됨)
import cv2
from   matplotlib import pyplot as plt

imageFile = './bono.jpg' #파일명
imgBGR = cv2.imread(imageFile) # cv2.IMREAD_COLOR(생략)
plt.axis('off') #눈금축 제거
plt.imshow(imgBGR)
plt.show()

imgRGB = cv2.cvtColor(imgBGR,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
plt.show()
```

```python
import cv2
from   matplotlib import pyplot as plt

imageFile = './bono.jpg'
imgGray = cv2.imread(imageFile, cv2.IMREAD_GRAYSCALE) #GRAYSCALE로 파일 읽기
plt.axis('off') #눈금축 제거

plt.imshow(imgGray, cmap = "gray", interpolation='bicubic')
#보간법: 몇개의 데이터들 사이의 값을 추정하는 방법
#보간법 종류:‘linear’, ‘nearest’, ‘zero’, ‘slinear’, ‘quadratic’, ‘cubic’, ‘previous’,
#              'next’, where ‘zero’, ‘slinear’, ‘quadratic’, ‘cubic',' bicubic'
plt.show()
```

```python
import cv2
import numpy as np

src = cv2.imread('C:/Users/Affinity/bono.jpg', cv2.IMREAD_COLOR) #그레이스케일(cv2.IMREAD_CRAYSCALE)
dst = np.zeros(src.shape, dtype=src.dtype)

N = 16 # 8, 32, 64
height, width,_ = src.shape # 컬러영상
##height, width = src.shape    # 그레이스케일 영상

h = height // N
w = width  // N
for i in range(N):
    for j in range(N):
        y = i*h
        x = j*w       
        roi = src[y:y+h, x:x+w]
        dst[y:y+h, x:x+w] = cv2.mean(roi)[0:3] # 컬러영상        
##        dst[y:y+h, x:x+w] = cv2.mean(roi)[0]   # 그레이스케일 영상

        
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
